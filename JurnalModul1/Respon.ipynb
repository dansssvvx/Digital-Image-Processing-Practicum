```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import math
```

```python
img = plt.imread('FORMAL.jpg')
plt.imshow(img)
```

### ROTASI

```python
tinggi, lebar = img.shape[:2]

rotasi = np.zeros_like(img)

theta = math.radians(-180)
center_x = lebar / 2
center_y = tinggi / 2

for y in range(tinggi):
    for x in range(lebar):
        x1 = int((x - center_x) * math.cos(theta) - (y - center_y) * math.sin(theta) + center_x)
        y1 = int((x - center_x) * math.sin(theta) + (y - center_y) * math.cos(theta) + center_y)
        if x1 >= 0 and x1 < lebar and y1 >= 0 and y1 < tinggi:
            rotasi[y][x] = img[y1][x1]

plt.imshow(rotasi, cmap='gray')
```

### FLIP

```python
flip = np.zeros_like(img)

for y in range(tinggi):
    for x in range(lebar):
        flip[y][x] = img[y][lebar-x-1]

plt.imshow(flip)
```

### TRANSLASI

```python
translasi = np.zeros_like(img)

for y in range(tinggi):
    for x in range(lebar):
        if (x + 163 < lebar and y - 300 > 0):
            translasi[y][x] = img[y-300][x+163]

plt.imshow(translasi)
```

### TRANSLASI

```python
dilatasi = np.zeros((tinggi * 3, lebar * 3, translasi.shape[2]), dtype=translasi.dtype)

for y in range(tinggi * 3):
    for x in range(lebar * 3):
        dilatasi[y][x] = translasi[int(y / 3)][int(x / 3)]

plt.imshow(dilatasi)
```
